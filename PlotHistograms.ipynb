```python

def rose_diagram(ax, vals, ymax, label, N=24, 
                 xticks=['Midnight', '3am', '6am', '9am', 'Midday', '3pm', '6pm', '9pm'],
                 wrapped_normal=None,
                 wrapped_normal_mixture=None, 
                 scalingFunc = np.sqrt):
    """
    Rose diagram, used for plotting Histogram over 24 hours (or other count)
    Note that the area in a segment is proportional to the count, 
    so the radius is proportional to the square root of the count.
    """
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
    width = (2*np.pi) / N
    
    radii = scalingFunc(vals)
    ymax  = scalingFunc(ymax)
    
    bars = ax.bar(theta, radii, width=width)
    ax.set_xticklabels(xticks)
    dtheta = 360 / len(xticks)
    #ax[:set_thetagrids]([0:dtheta:360-dtheta])
    #ax[:set_thetagrids]([0:dtheta:(2:np.pi) - dtheta])
    ax.set_thetagrids(np.arange(0,360,dtheta))
    
    # Set y limit to the square root of the max, and set the tick labels to be 
    # the values squared. Also only show every third tick label
    ax.set_ylim(0,(ymax))
    ticks = ax.get_yticks()[::2]
    ax.set_yticks(ticks)
    ax.set_yticklabels([int(t) for t in np.power(ticks,2)])
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    
    ret = None
    
    if wrapped_normal is not None:
        rng = np.linspace(0, wrapped_normal._period, 100, endpoint=True)
        pdf = wrapped_normal.pdf(rng)
        pdfs = pdf / max(pdf) * (ymax)
        ax.plot(rng / wrapped_normal._period * 2 * np.pi, pdfs, 'r', linewidth=2)
        label += "\n" + str(wrapped_normal)
        
        ret = pdfs
    
    if wrapped_normal_mixture is not None:
        rng = np.linspace(0, wrapped_normal_mixture._period, 100, endpoint=True)
        mixtureSums = []
        for wn in wrapped_normal_mixture._wrapped_normals:
            pdf = np.maximum(0, wn.pdf(rng))
            pdfs = pdf / max(pdf) * np.sqrt(wn._weight) * (ymax)
            
            mixtureSums.append( np.asarray( pdfs ) )
            label += "\n" + str(wn)
        
        mixtureSum = np.asarray( mixtureSums ).sum( 0 )
        
        for mm in mixtureSums:
            ax.plot(rng / wn._period * 2 * np.pi, mm / mixtureSum.max(), 'r', linewidth=1)
        
        ax.plot(rng / wn._period * 2 * np.pi, mixtureSum / mixtureSum.max(), 'k', linewidth=2)
        label += "\n"
        
        ret = mixtureSum 
        
            
    ax.set_title(label, y=1.08)
    
    return ret
```

```python
class WrappedNormal(object):
    def __init__(self, mean, variance, period=2*np.pi, weight=1.0):
        self._mean = mean
        self._variance = variance
        self._period = period
        self._weight = weight
        
    def __repr__(self):
        return 'WrappedNormal(mean={0}, variance={1}, period={2})'.format(self._mean, self._variance, self._period)
        
    def __str__(self):
        label = 'WN({0:.2f},{1:.2f},{2:0.1f}'.format(self._mean, self._variance, self._period)
        if self._weight != 1.0:
            label = label + ',{0:0.1f}'.format(self._weight)
        return label + ')'
    
    def pdf(self, x, verbose=False):
            # p(x) = sum_k N(x + L*k; m, v)
            #      = a/(2*pi) + a/pi sum_{k=1}^inf cos(a*k*(x-m)) exp(-(a*k)^2 v/2) where a = 2*pi/L
            if self._variance < 0.15 * pow(self._period / 2 / np.pi, 2):
                # use Gaussian summation formula
                if verbose:
                    print('Using Gaussian summation formula')
                result = 0 #-np.infty;
                for k in xrange(-1,1):
                    # logp = Gaussian.GetLogProb(value + k*Period)
                    # result = MMath.LogSumExp(result, logp)
                    tmp = norm.pdf(x + k*self._period, self._mean, self._variance)
                    #print(tmp.min(), tmp.max())
                    result += tmp
                return result
            else:
                if verbose:
                    print('Using cosine formula')
                # v >= 0.15
                # use the cosine formula
                result = 0.5
                aOverPi = 2.0 / self._period
                a = aOverPi * np.pi
                diff = x - self._mean
                vHalf = self._variance * 0.5
                for k in xrange(1,8):
                    ak = a*k
                    result += np.cos(ak*diff) * np.exp(-ak*ak*vHalf)
                #print(result, aOverPi, self._period)
                return result * aOverPi

class WrappedNormalMixture(object):
    def __init__(self, wrapped_normals, weights):
        self._period = wrapped_normals[0]._period
        for wn in wrapped_normals[1:]:
            if wn._period != self._period:
                # BOMB OUT HERE
                raise error('Periods must be the same')
        self._wrapped_normals = wrapped_normals
        for (weight, wn) in zip(weights, self._wrapped_normals):
            wn._weight = weight
        self._weights = weights
        
```

```python
def plotRoseDiagrams( path, filename, quants = 1, scalingFunc = lambda xx: xx ):
    bins = np.linspace( 0, 24, 24 * quants + 1 )
    
    
    dataParams = json.load( open( path + "/" + filename, "r" ) )
    
    items = dataParams["Params"]
    data = np.asarray( dataParams["Data"] )
    
    for ii in np.arange( len( items ) ): 
        while items[ii]["Mean"] < 0:
            items[ii]["Mean"] += 24
            
        while items[ii]["Mean"] > 24: 
            items[ii]["Mean"] -= 24
            
        #print (   "Mean:     {Mean}" + 
        #        "\nVariance: {Variance} (Precision: {Precision})" + 
        #        "\nWeight:   {Weight}\n" ).format( **items[ii] )
    
    df = pd.DataFrame( items ) 
    print df 
        
    hist = np.histogram( data, bins = bins )
    hist = np.asarray( hist[0], dtype = np.float )
    hist /= hist.max()
    
    wns = []
    weights = []
    
    for item in items:
        weights.append( item['Weight'] )
        wns.append( WrappedNormal( item['Mean'], item['Variance'], item['Period'] ) )
        
    pl.figure( figsize=(15,6) )
    ax = pl.subplot( 1, 2, 1, polar=True )
    wnm = WrappedNormalMixture( wns, weights )
    likelihood = rose_diagram( ax, hist, hist.max(), "ads", N = len( bins ) - 1, 
                              wrapped_normal_mixture = wnm, scalingFunc=scalingFunc )
    
    
    pl.subplot( 1, 2, 2 ) 
    a = pl.hist( data, bins = bins );
    pl.plot( np.linspace( 0, 24, len( likelihood ) ), 0.7*a[0].max() * likelihood / likelihood.max(), 'k', linewidth=2 )
    pl.xlim( ( 0, 24 ) )
```
